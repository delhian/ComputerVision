# Задание 1
Нужно локализовать на изображении дорожный знак пешеходный переход с помощью цветового порога. Результатом у вас должна получиться маска изображения, на которой хорошо выделен дорожный знак. (Рекомендую перевести изображение в HSV и погуглить функцию inRange). Финальную маску изображения попрошу назвать result_image.

```python
pesh = cv2.imread('peshehod.jpeg')
pesh = cv2.cvtColor(pesh, cv2.COLOR_BGR2RGB)
plt.imshow(pesh) 
```

```python
pesh_hsv = cv2.cvtColor(pesh, cv2.COLOR_BGR2HSV)

print('Hue', pesh_hsv[:,:, 0].min(), pesh_hsv[:,:, 0].max())
print('Saturation', pesh_hsv[:,:, 1].min(), pesh_hsv[:,:, 1].max())
print('Value/Brightness', pesh_hsv[:,:, 2].min(), pesh_hsv[:,:, 2].max())
```

```python
result_image = cv2.inRange(pesh_hsv, (0, 180, 0), (179, 255, 255))
plt.imshow(result_image, cmap='gray')
np.sum(result_image[120:250, 120:250])/np.sum(result_image)
```

# Задание 2
Для изображения знака сделайте threshold по порогу выше 180.

```python
gray_pesh = cv2.cvtColor(pesh, cv2.COLOR_RGB2GRAY)
_, thresholded_pesh = cv2.threshold(gray_pesh, 180, 255, 0)
plt.subplot(221)
plt.imshow(gray_pesh)
plt.subplot(222)
plt.imshow(thresholded_pesh)
plt.show()


print(np.sum(thresholded_pesh)/255)
```
